```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import to_rgb

df = pd.read_csv('means_df.csv')

x_coords = np.array([
    0.500, 2.500, 5, 7.500, 10, 13, 16, 19, 22, 28, 35, 42, 52, 65, 79.33, 98, 115, 135, 0.500,
    3.500, 7, 11, 15, 20, 25, 40, 55, 70, 85, 100, 115, 135
]) / 140  # Skalierung

tikz_code_upper = ""
tikz_code_lower = ""
num_aoa = len(df.columns)

colors = plt.cm.viridis(np.linspace(0, 1, num_aoa))

for idx, aoa in enumerate(df.columns):
    y_values_upper = df.iloc[:18, idx].values
    y_values_lower = df.iloc[18:, idx].values
    
    color = colors[idx]
    rgb_color = to_rgb(color)
    tikz_color = f"{{rgb,255:red,{int(rgb_color[0]*255)};green,{int(rgb_color[1]*255)};blue,{int(rgb_color[2]*255)}}}"

    tikz_code_upper += f"\\addplot[thick, color={tikz_color}, mark=*] coordinates {{\n"
    for x, y in zip(x_coords[:18], y_values_upper):
        tikz_code_upper += f"({x}, {y})\n"
    tikz_code_upper += "};\n"
    
    tikz_code_lower += f"\\addplot[thick, color={tikz_color}, mark=*, dashed] coordinates {{\n"
    for x, y in zip(x_coords[18:], y_values_lower):
        tikz_code_lower += f"({x}, {y})\n"
    tikz_code_lower += "};\n"

print("Oberseite TikZ Code:\n", tikz_code_upper)
print("Unterseite TikZ Code:\n", tikz_code_lower)

```
