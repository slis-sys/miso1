```python
path = './flow/Stat_20240531_BL_180_10_01'
mat_files = [f for f in os.listdir(path) if f.endswith('.mat')]
means = np.empty((40,21))
rio = np.empty((3,21))
i = 0

for f in tqdm(mat_files):
    aoa_value = float(f[-10:-4])
    
    if 0.0 <= aoa_value <= 180:
        filepath = os.path.join(path, f)
        f1 = loadmat(filepath)
        
        if f.startswith('Stat_20240531_BL_180_10_01_cDAQ'):
            for j, (key, value) in enumerate(f1.items()):
                if key.startswith('cDAQ') and j < 40:
                    means[j, i] = np.mean(value)
        elif f.startswith('Stat_20240531_BL_180_10_01_RIO'):
            k = 0
            for key, value in f1.items():
                if key.startswith('AI00'):
                    rio[k, i] = np.mean(value)
                    k += 1
                elif key.startswith('Dens') or key.startswith('Sam'):
                    rio[k, i] = float(value)
                    k += 1
                if k == 3:
                    break
            i += 1
```
